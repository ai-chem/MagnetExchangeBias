Pauling electronegativity (χ)

```python
# Функция для расчета атомного процента и суммы произведений атомного процента на χ
def calculate_properties(material, element_data):
    try:
        comp = Composition(material)
        total_atoms = comp.num_atoms  # Общее количество атомов в формуле

        # Считаем атомные проценты и произведения на χ
        atomic_percents = {el: (amt / total_atoms) * 100 for el, amt in comp.get_el_amt_dict().items()}
        sum_chi_c = sum(atomic_percents[el] * element_data[el]['Electronegativity (χ)'] for el in atomic_percents if el in element_data)

        return atomic_percents, sum_chi_c
    except ValueError as e:
        print(f"Error processing formula {material}: {e}")
        return None, None
```

```python
# Функция для расчета суммы произведений атомного процента на VEC
def calculate_sum_vec(atomic_percents, element_data):
    return sum(atomic_percents.get(el, 0) * element_data[el]['VEC'] for el in atomic_percents if el in element_data)

# Функция для расчета суммы произведений атомного процента на Work function (W)
def calculate_sum_work_function(atomic_percents, element_data):
    return sum(atomic_percents.get(el, 0) * (element_data[el]['Work function (W)'] if element_data[el]['Work function (W)'] is not None else 0) for el in atomic_percents if el in element_data)

# Функция для расчета суммы произведений атомного процента на χ
def calculate_sum_chi_c(atomic_percents, element_data):
    return sum(atomic_percents.get(el, 0) * element_data[el]['Electronegativity (χ)'] for el in atomic_percents if el in element_data)
# Функция для получения атомных процентов
def get_atomic_percents(material, element_data):
    try:
        comp = Composition(material)
        total_atoms = comp.num_atoms
        atomic_percents = {el: (amt / total_atoms) * 100 for el, amt in comp.get_el_amt_dict().items()}
        return atomic_percents
    except ValueError as e:
        print(f"Error processing formula {material}: {e}")
        return {}
```

Valence electron concentration (VEC)

```python
# Функция для расчета суммы произведений атомного процента на VEC
def calculate_sum_vec(atomic_percents, element_data):
    return sum(atomic_percents.get(el, 0) * element_data[el]['VEC'] for el in atomic_percents if el in element_data)

# Функция для расчета суммы произведений атомного процента на Work function (W)
def calculate_sum_work_function(atomic_percents, element_data):
    return sum(atomic_percents.get(el, 0) * (element_data[el]['Work function (W)'] if element_data[el]['Work function (W)'] is not None else 0) for el in atomic_percents if el in element_data)

# Функция для получения атомных процентов
def get_atomic_percents(material, element_data):
    try:
        comp = Composition(material)
        total_atoms = comp.num_atoms
        atomic_percents = {el: (amt / total_atoms) * 100 for el, amt in comp.get_el_amt_dict().items()}
        return atomic_percents
    except ValueError as e:
        print(f"Error processing formula {material}: {e}")
        return {}

# Применяем функции к столбцам DataFrame
df['core_atomic_percent'] = df['core'].apply(lambda x: get_atomic_percents(x, element_data))
df['shell_atomic_percent'] = df['shell'].apply(lambda x: get_atomic_percents(x, element_data))

# Рассчитываем сумму произведений для VEC и Work function (W)
df['core_sum_vec'] = df['core_atomic_percent'].apply(lambda x: calculate_sum_vec(x, element_data))
df['core_sum_work_function'] = df['core_atomic_percent'].apply(lambda x: calculate_sum_work_function(x, element_data))

df['shell_sum_vec'] = df['shell_atomic_percent'].apply(lambda x: calculate_sum_vec(x, element_data))
df['shell_sum_work_function'] = df['shell_atomic_percent'].apply(lambda x: calculate_sum_work_function(x, element_data))

# Выводим результат
print(df)
```

```python
# Функция для расчета средних значений для материала с обработкой None значений
def calculate_mean_property(atomic_percents, element_data, property_name):
    total = 0
    for el in atomic_percents:
        if el in element_data:
            property_value = element_data[el].get(property_name)
            if property_value is not None:
                total += atomic_percents[el] * property_value
    return total / 100  # делим на 100, т.к. атомные проценты суммируются до 100%

# Получаем средние значения для электроотрицательности, VEC и работы выхода
df['core_mean_electronegativity'] = df['core_atomic_percent'].apply(lambda x: calculate_mean_property(x, element_data, 'Electronegativity (χ)'))
df['core_mean_vec'] = df['core_atomic_percent'].apply(lambda x: calculate_mean_property(x, element_data, 'VEC'))
df['core_mean_work_function'] = df['core_atomic_percent'].apply(lambda x: calculate_mean_property(x, element_data, 'Work function (W)'))

df['shell_mean_electronegativity'] = df['shell_atomic_percent'].apply(lambda x: calculate_mean_property(x, element_data, 'Electronegativity (χ)'))
df['shell_mean_vec'] = df['shell_atomic_percent'].apply(lambda x: calculate_mean_property(x, element_data, 'VEC'))
df['shell_mean_work_function'] = df['shell_atomic_percent'].apply(lambda x: calculate_mean_property(x, element_data, 'Work function (W)'))
```

```python
# Функция для расчета разности для Pauling Electronegativity Difference
def calculate_electronegativity_difference(atomic_percents, element_data, mean_electronegativity):
    if mean_electronegativity == 0:
        return 0
    return np.sqrt(sum(atomic_percents.get(el, 0) * ((1 - element_data[el]['Electronegativity (χ)'] / mean_electronegativity) ** 2)
                      for el in atomic_percents if el in element_data and element_data[el]['Electronegativity (χ)'] is not None))

# Функция для расчета разности для Valence Electron Concentration Difference
def calculate_vec_difference(atomic_percents, element_data, mean_vec):
    if mean_vec == 0:
        return 0
    return np.sqrt(sum(atomic_percents.get(el, 0) * ((1 - element_data[el]['VEC'] / mean_vec) ** 2)
                      for el in atomic_percents if el in element_data and element_data[el]['VEC'] is not None))

# Функция для расчета разности для Work Function Difference
def calculate_work_function_difference(atomic_percents, element_data, mean_work_function):
    if mean_work_function == 0:
        return 0
    return np.sqrt(sum(atomic_percents.get(el, 0) * ((1 - (element_data[el]['Work function (W)'] if element_data[el]['Work function (W)'] is not None else 0) / mean_work_function) ** 2)
                      for el in atomic_percents if el in element_data))
```

```python
# Рассчитываем разности для core
df['core_pauling_electronegativity_difference'] = df.apply(lambda row: calculate_electronegativity_difference(row['core_atomic_percent'], element_data, row['core_mean_electronegativity']), axis=1)
df['core_vec_difference'] = df.apply(lambda row: calculate_vec_difference(row['core_atomic_percent'], element_data, row['core_mean_vec']), axis=1)
df['core_work_function_difference'] = df.apply(lambda row: calculate_work_function_difference(row['core_atomic_percent'], element_data, row['core_mean_work_function']), axis=1)

# Рассчитываем разности для shell
df['shell_pauling_electronegativity_difference'] = df.apply(lambda row: calculate_electronegativity_difference(row['shell_atomic_percent'], element_data, row['shell_mean_electronegativity']), axis=1)
df['shell_vec_difference'] = df.apply(lambda row: calculate_vec_difference(row['shell_atomic_percent'], element_data, row['shell_mean_vec']), axis=1)
df['shell_work_function_difference'] = df.apply(lambda row: calculate_work_function_difference(row['shell_atomic_percent'], element_data, row['shell_mean_work_function']), axis=1)
```

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.core.composition import Composition
import pandas as pd

# Ваш API ключ
api_key = "dS2WjhWZea9vufDmKOGHiJOfOhbIZ4B9"

def get_integer_formula(material):
    try:
        comp = Composition(material)
        integer_formula, factor = comp.get_integer_formula_and_factor()
        return integer_formula
    except ValueError as e:
        print(f"Error processing formula {material}: {e}")
        return None

def get_reduced_formula(material):
    try:
        comp = Composition(material)
        reduced_formula, factor = comp.get_reduced_formula_and_factor()
        return reduced_formula
    except ValueError as e:
        print(f"Error processing formula {material}: {e}")
        return None

# Преобразование формул в core и shell
df["core"] = df["core"].apply(lambda x: get_reduced_formula(get_integer_formula(x)) if pd.notna(x) and x not in ["0", "", " "] else x)
df["shell"] = df["shell"].apply(lambda x: get_reduced_formula(get_integer_formula(x)) if pd.notna(x) and x not in ["0", "", " "] else x)

# Инициализация MPRester с использованием вашего API ключа
with MPRester(api_key) as m:
    def get_lattice_params(material):
        if pd.isna(material) or material in ["0", "", " "]:
            return None, None, None, None, None, None
        try:
            # Получение структурных данных для материала
            entries = m.get_entries(material, inc_structure="final")

            # Фильтрация, чтобы выбрать наиболее компактную структуру
            min_atoms = float('inf')
            best_entry = None
            for entry in entries:
                num_atoms = len(entry.structure)
                if num_atoms < min_atoms:
                    min_atoms = num_atoms
                    best_entry = entry

            # Извлечение параметров решетки и углов
            if best_entry:
                lattice = best_entry.structure.lattice
                a, b, c = lattice.a, lattice.b, lattice.c
                alpha, beta, gamma = lattice.alpha, lattice.beta, lattice.gamma
                return a, b, c, alpha, beta, gamma
            else:
                return None, None, None, None, None, None
        except Exception as e:
            print(f"Error retrieving data for material {material}: {e}")
            return None, None, None, None, None, None

    # Создание столбцов для хранения параметров решетки и углов
    df["core_a"], df["core_b"], df["core_c"] = None, None, None
    df["core_alpha"], df["core_beta"], df["core_gamma"] = None, None, None
    df["shell_a"], df["shell_b"], df["shell_c"] = None, None, None
    df["shell_alpha"], df["shell_beta"], df["shell_gamma"] = None, None, None

    # Извлечение и сохранение параметров для каждого материала в core и shell
    for index, row in df.iterrows():
        core = row["core"]
        shell = row["shell"]

        core_params = get_lattice_params(core)
        shell_params = get_lattice_params(shell)

        if core_params:
            df.at[index, "core_a"], df.at[index, "core_b"], df.at[index, "core_c"] = core_params[:3]
            df.at[index, "core_alpha"], df.at[index, "core_beta"], df.at[index, "core_gamma"] = core_params[3:]

        if shell_params and shell not in ["0", "", " "]:
            df.at[index, "shell_a"], df.at[index, "shell_b"], df.at[index, "shell_c"] = shell_params[:3]
            df.at[index, "shell_alpha"], df.at[index, "shell_beta"], df.at[index, "shell_gamma"] = shell_params[3:]

print(df)
```

# Ищем магнитные ионы

```python
# Задаем ионы, которые считаются магнитными
magnetic_ions = ["Fe", "Co", "Ni", "Mn", "Cr", "V", "Zn", "Nd", "Sm", "Dy", "Gd", "Ru", "Ti", "Cu", "Ag", "La"]

def count_ions(formula):
    if pd.isna(formula) or formula in ["0", "", " ", None]:
        return 0, 0
    # Создаем объект Composition для данной формулы
    try:
        composition = Composition(formula)
        total_ions = sum(composition[el] for el in composition)
        magnetic_ions_count = sum(composition[el] for el in composition if el.symbol in magnetic_ions)
    except ValueError:
        total_ions, magnetic_ions_count = 0, 0
    return total_ions, magnetic_ions_count

# Функция для вычисления суммы всех ионов и суммы магнитных ионов для каждой строки
def calculate_ions(row):
    core_total_ions, core_magnetic_ions = count_ions(row["core"])
    shell_total_ions, shell_magnetic_ions = count_ions(row["shell"])
    total_ions = core_total_ions + shell_total_ions
    magnetic_ions = core_magnetic_ions + shell_magnetic_ions
    return pd.Series({
        "total_ions": total_ions,
        "z_ions": magnetic_ions,
        "core_total_ions": core_total_ions,
        "core_magnetic_ions": core_magnetic_ions,
        "shell_total_ions": shell_total_ions,
        "shell_magnetic_ions": shell_magnetic_ions
    })
```

```python
# Применяем функцию к каждой строке и сохраняем результат в новые столбцы
df[["total_ions", "z_ions", "core_total_ions", "core_magnetic_ions", "shell_total_ions", "shell_magnetic_ions"]] = df.apply(calculate_ions, axis=1)

print(df)
```

# Aex= (3*Tc)/(2*z*a)
a = lattice constant
z = number of magnetic ions

```python
df = pd.read_csv(r'/content/bias_param_0508_ - bias_param_0508_.csv (1).csv')
```

```python
#df['Tc'] = df['Tc'].fillna(0)
#df['shell_a'] = df['shell_a'].fillna(0)

# Функция для вычисления Aex
def calculate_Aex(row):
    try:
        Tc = float(row["Tc"])  # Приведение к float
        z_ions = float(row["z_ions"])  # Приведение к float
        shell_a = float(row["shell_a"])  # Приведение к float
        core_a = float(row["core_a"])  # Приведение к float

        if z_ions != 0:
            if shell_a != 0:
                Aex = (3 * Tc) / (shell_a + core_a)/2
            elif shell_a == 0:
                Aex = (3 * Tc) / (2 * z_ions * core_a)
            else:
                Aex = np.nan  # Если оба shell_a и core_a равны нулю
        else:
            Aex = np.nan  # Если z_ions равно нулю
    except (ValueError, TypeError):
        Aex = np.nan  # На случай непредвиденных ошибок
    return Aex

# Создаем столбец Aex
df["Aex"] = df.apply(calculate_Aex, axis=1)

print(df)
```
